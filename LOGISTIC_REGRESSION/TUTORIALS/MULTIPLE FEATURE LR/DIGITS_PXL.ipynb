```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits as LG
```

```python
digits = LG()
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range (5):

    plt.matshow(digits.images[i])
```

```python
array_format = []
for i in range(5):
    print(digits.target[i])
    print(type(digits.target[i]))
    array_format.append(digits.target[i])

array_format
```

```python
digits.target[0:5]
type(digits.target[0:5])
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test =tts(digits.data,digits.target,test_size=0.2,random_state=10)
 
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression as LR
model=LR()
model.fit(x_train,y_train)  #training the model
 
```

```python
model.score(x_test,y_test)  #testing the model
```

```python
plt.matshow(digits.images[67])
```

```python
digits.data[0:5]
```

```python
model.predict(digits.data[0:5])
```

```python
from sklearn.metrics import confusion_matrix as cm
y_predicted=model.predict(x_test)
confusion_mat=cm(y_test,y_predicted)
confusion_mat
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(confusion_mat,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.title('Confusion Matrix')
plt.show()
```
