```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris.keys()
```

```python
for i in range(len(iris.data)):
    print(f"Sample {i+1}: {iris.data[i]}")
    for j in range(len(iris.feature_names)):
        print(f"\t{iris.feature_names[j]}: {iris.data[i][j]}")
```

```python
iris.target
```

```python
# Map the target array to the corresponding flower names
flower_names = [iris.target_names[target] for target in iris.target]
flower_names

```

```python
iris.target_names
```

```python
import pandas as pd
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head(30)
```

```python
df['Target_flower_name']=flower_names
df.head(80)
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression

x = df[iris.feature_names]
y = df['Target_flower_name']
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
Predicted_flower_names = model.predict(x_test)
mod_score=accuracy_score(y_test, Predicted_flower_names)
mod_score
```

```python
new_data=[[5.1, 3.5, 1.4, 0.2]]
new_pred=model.predict(new_data)
new_pred[0]
```
